```python
import numpy as np
import pandas as pd
from pandas import DataFrame
from pandas import Series
import seaborn as sns
from tqdm import tqdm
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
font_path = 'AppleGothic.ttf'
fontprop = fm.FontProperties(fname=font_path, size=18)
font_name = fm.FontProperties(fname=font_path).get_name()
import statsmodels.api as sm
from math import *
```

```python
weather = pd.read_csv("제주기상자료(월단위).csv", encoding="cp949")
```

```python
weather.head()
```

```python
temp_1 = weather[weather['지점명']=='제주']
temp_1 = temp_1.drop(['지점','평균최고기온(°C)','평균최저기온(°C)','평균기온(°C)'],axis=1, inplace=False)
temp_1 = temp_1.reset_index(drop=False)
temp_1['월합강수량(00~24h만)(mm)']= np.log(temp_1['월합강수량(00~24h만)(mm)']).diff()
temp_1= temp_1.dropna()

temp_1['sinx/6']= temp_1['index'].apply(lambda x: sin((x+1)*pi/6))
temp_1['cosx/6']= temp_1['index'].apply(lambda x: cos((x+1)*pi/6))
temp_1['sinx/3']= temp_1['index'].apply(lambda x: sin((x+1)*pi/3))
temp_1['cosx/3']= temp_1['index'].apply(lambda x: cos((x+1)*pi/3))
temp_1['sinx/2']= temp_1['index'].apply(lambda x: sin((x+1)*pi/2))
temp_1['cosx/2']= temp_1['index'].apply(lambda x: cos((x+1)*pi/2))
temp_1['sinx/12']= temp_1['index'].apply(lambda x: sin((x+1)*pi/12))
temp_1['cosx/12']= temp_1['index'].apply(lambda x: cos((x+1)*pi/12))
temp_1['sinx/4']= temp_1['index'].apply(lambda x: sin((x+1)*pi/4))
temp_1['cosx/4']= temp_1['index'].apply(lambda x: cos((x+1)*pi/4))
```

```python
temp_1.head()
```

```python
target = temp_1['월합강수량(00~24h만)(mm)']
x_data_1 = temp_1[['sinx/6','cosx/6','sinx/3','cosx/3','sinx/12','sinx/2','cosx/2','cosx/12','sinx/4','cosx/4']]
```

```python
x_data_1
```

```python
x_data_1  = sm.add_constant(x_data_1  , has_constant = "add")
```

```python
multi_model = sm.OLS(target, x_data_1)
fitted_multi_model = multi_model.fit()

fitted_multi_model.summary()
```
