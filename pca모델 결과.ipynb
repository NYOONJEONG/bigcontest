```python
resd_prep = pd.DataFrame()

for i in range(1, 10):
    temp_data = temp2_resd[temp2_resd['emd_cluster']==i]
    for j in set(temp_data['emd_nm']):
        temp_data_1 = temp_data[temp_data['emd_nm'] == j]
        for k in set(temp_data_1['age_cluster_resd']):
            temp_data_2 = temp_data_1[temp_data_1['age_cluster_resd'] == k]
            for l in set(temp_data_2['age_time_cluster_resd']):
                temp_data_3 = temp_data_2[temp_data_2['age_time_cluster_resd'] == l]
                temp_method = method_resd[method_resd['cluster'] == str(i)+'-'+str(k)+'-'+str(l)]
                if not(temp_method['method'].iloc[0] in ['diff', 'log', 'log_diff']):
                    continue
                elif (temp_method['method'].iloc[0] == 'diff'):
                    temp_data_3['resd_pop_cnt'] = temp_data_3['resd_pop_cnt'].diff().shift(1)
                    
                elif (temp_method['method'].iloc[0] == 'log'):
                    temp_data_3['resd_pop_cnt'] = np.log(temp_data_3['resd_pop_cnt']).shift(1)
                    
                elif (temp_method['method'].iloc[0] == 'log_diff'):
                    temp_data_3['resd_pop_cnt'] = np.log(temp_data_3['resd_pop_cnt']).diff().shift(1)
                resd_prep = pd.concat([resd_prep, temp_data_3.reset_index(drop=True)], axis = 0)
                
resd_prep = resd_prep.sort_values(by=['emd_cluster', 'emd_nm', 'age_cluster_resd','age_time_cluster_resd','year_month']).reset_index(drop=True)
```

```python
import statsmodels.api as sm
temp_result1 = pd.DataFrame(index=sorted(set(data1['emd_nm'])), columns=['2021-07', '2021-08'])
for i in sorted(set(data1['emd_nm'])):
    data1_temp = data1[data1['emd_nm'] == i].reset_index(drop=True)
    data1_X = data1_temp.drop(['cluster', 'emd_nm', 'year_month', 'em_g'], axis=1).reset_index(drop=True)
    data1_y = data1_temp['em_g'].dropna().reset_index(drop=True)
    data1_X = sm.add_constant(data1_X, has_constant = 'add')
    predict1_X = pd.DataFrame(np.mean(data1_X[data1_X.columns].iloc[[-14,-26,-38],:])).T
    predict2_X = pd.DataFrame(np.mean(data1_X[data1_X.columns].iloc[[-13,-25,-37],:])).T
    data1_X = data1_X.dropna()
    data1_y = data1_temp.dropna()['em_g']
    
    temp_loop=True
    while (temp_loop):
        model = sm.OLS(data1_y, data1_X)
        fitted_model = model.fit()
        p = pd.DataFrame(fitted_model.pvalues).reset_index()
        if (p.iloc[1:,1] > 0.05).sum() == 0:
            temp_loop = False
        else:
            temp_drop = p[p.iloc[0:,1] == max(p.iloc[1:,1])].iloc[0]['index']
            data1_X = data1_X.drop([temp_drop], axis=1)
    model = sm.OLS(data1_y, data1_X)
    fitted_model = model.fit()
    print(i)
    print(data1_X.columns)
    print(fitted_model.summary())
    print(fitted_model.predict(predict1_X[data1_X.columns]))
    print(fitted_model.predict(predict2_X[data1_X.columns]))
    temp_result1.loc[i, '2021-07'] = fitted_model.predict(predict1_X[data1_X.columns])[0]
    temp_result1.loc[i, '2021-08'] = fitted_model.predict(predict2_X[data1_X.columns])[0]
```
