```python
temp2 = data2.copy()
temp2 = temp2.drop(['Unnamed: 0', 'base_date', 'time', 'city', 'emd_cd', 'sex', 'age', 'time'], axis=1)
temp2_resd = temp2.groupby(['emd_cluster', 'emd_nm', 'age_cluster_resd', 'age_time_cluster_resd', 'year_month']).sum().reset_index()
temp2_work = temp2.groupby(['emd_cluster', 'emd_nm', 'age_cluster_work', 'age_time_cluster_work', 'year_month']).sum().reset_index()
temp2_visit = temp2.groupby(['emd_cluster', 'emd_nm', 'age_cluster_visit', 'age_time_cluster_visit', 'year_month']).sum().reset_index()
```

```python
temp2_resd = temp2_resd[['emd_cluster', 'emd_nm', 'age_cluster_resd','age_time_cluster_resd', 'year_month', 'resd_pop_cnt']]
temp2_work = temp2_work[['emd_cluster', 'emd_nm', 'age_cluster_work','age_time_cluster_work', 'year_month', 'work_pop_cnt']]
temp2_visit = temp2_visit[['emd_cluster', 'emd_nm', 'age_cluster_visit','age_time_cluster_visit', 'year_month', 'visit_pop_cnt']]
```

```python
resd_prep = pd.DataFrame()

for i in range(1, 10):
    temp_data = temp2_resd[temp2_resd['emd_cluster']==i]
    for j in set(temp_data['emd_nm']):
        temp_data_1 = temp_data[temp_data['emd_nm'] == j]
        for k in set(temp_data_1['age_cluster_resd']):
            temp_data_2 = temp_data_1[temp_data_1['age_cluster_resd'] == k]
            for l in set(temp_data_2['age_time_cluster_resd']):
                temp_data_3 = temp_data_2[temp_data_2['age_time_cluster_resd'] == l]
                temp_method = method_resd[method_resd['cluster'] == str(i)+'-'+str(k)+'-'+str(l)]
                if not(temp_method['method'].iloc[0] in ['diff', 'log', 'log_diff']):
                    continue
                elif (temp_method['method'].iloc[0] == 'diff'):
                    temp_data_3['resd_pop_cnt'] = temp_data_3['resd_pop_cnt'].diff().shift(1)
                    
                elif (temp_method['method'].iloc[0] == 'log'):
                    temp_data_3['resd_pop_cnt'] = np.log(temp_data_3['resd_pop_cnt']).shift(1)
                    
                elif (temp_method['method'].iloc[0] == 'log_diff'):
                    temp_data_3['resd_pop_cnt'] = np.log(temp_data_3['resd_pop_cnt']).diff().shift(1)
                resd_prep = pd.concat([resd_prep, temp_data_3.reset_index(drop=True)], axis = 0)
                
resd_prep = resd_prep.sort_values(by=['emd_cluster', 'emd_nm', 'age_cluster_resd','age_time_cluster_resd','year_month']).reset_index(drop=True)
```

```python
work_prep = pd.DataFrame()

for i in range(1, 10):
    temp_data = temp2_work[temp2_work['emd_cluster']==i]
    for j in set(temp_data['emd_nm']):
        temp_data_1 = temp_data[temp_data['emd_nm'] == j]
        for k in set(temp_data_1['age_cluster_work']):
            temp_data_2 = temp_data_1[temp_data_1['age_cluster_work'] == k]
            for l in set(temp_data_2['age_time_cluster_work']):
                temp_data_3 = temp_data_2[temp_data_2['age_time_cluster_work'] == l]
                temp_method = method_work[method_work['cluster'] == str(i)+'-'+str(k)+'-'+str(l)]
                if not(temp_method['method'].iloc[0] in ['diff', 'log', 'log_diff']):
                    continue
                elif (temp_method['method'].iloc[0] == 'diff'):
                    temp_data_3['work_pop_cnt'] = temp_data_3['work_pop_cnt'].diff().shift(1)
                    
                elif (temp_method['method'].iloc[0] == 'log'):
                    temp_data_3['work_pop_cnt'] = np.log(temp_data_3['work_pop_cnt']).shift(1)
                    
                elif (temp_method['method'].iloc[0] == 'log_diff'):
                    temp_data_3['work_pop_cnt'] = np.log(temp_data_3['work_pop_cnt']).diff().shift(1)
                work_prep = pd.concat([work_prep, temp_data_3.reset_index(drop=True)], axis = 0)
                
work_prep = work_prep.sort_values(by=['emd_cluster', 'emd_nm', 'age_cluster_work','age_time_cluster_work','year_month']).reset_index(drop=True)
```

```python
visit_prep = pd.DataFrame()

for i in range(1, 10):
    temp_data = temp2_visit[temp2_visit['emd_cluster']==i]
    for j in set(temp_data['emd_nm']):
        temp_data_1 = temp_data[temp_data['emd_nm'] == j]
        for k in set(temp_data_1['age_cluster_visit']):
            temp_data_2 = temp_data_1[temp_data_1['age_cluster_visit'] == k]
            for l in set(temp_data_2['age_time_cluster_visit']):
                temp_data_3 = temp_data_2[temp_data_2['age_time_cluster_visit'] == l]
                temp_method = method_visit[method_visit['cluster'] == str(i)+'-'+str(k)+'-'+str(l)]
                if not(temp_method['method'].iloc[0] in ['diff', 'log', 'log_diff']):
                    continue
                elif (temp_method['method'].iloc[0] == 'diff'):
                    temp_data_3['visit_pop_cnt'] = temp_data_3['visit_pop_cnt'].diff().shift(1)
                    
                elif (temp_method['method'].iloc[0] == 'log'):
                    temp_data_3['visit_pop_cnt'] = np.log(temp_data_3['visit_pop_cnt']).shift(1)
                    
                elif (temp_method['method'].iloc[0] == 'log_diff'):
                    temp_data_3['visit_pop_cnt'] = np.log(temp_data_3['visit_pop_cnt']).diff().shift(1)
                visit_prep = pd.concat([visit_prep, temp_data_3.reset_index(drop=True)], axis = 0)
                
visit_prep = visit_prep.sort_values(by=['emd_cluster', 'emd_nm', 'age_cluster_visit','age_time_cluster_visit','year_month']).reset_index(drop=True)
```
