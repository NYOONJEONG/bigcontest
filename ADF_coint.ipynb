```python
import pandas as pd
data1 = pd.read_csv('./01_제공데이터_update(210811)/01_음식물쓰레기_FOOD_WASTE_210811_update.csv', encoding='cp949')
data2 = pd.read_csv('./유동인구_clustering_final.csv')
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
def get_year_month(x):
    return x[0:7]
```

```python
data1['year_month'] = data1['base_date'].apply(lambda x: get_year_month(x))
data2['year_month'] = data2['base_date'].apply(lambda x: get_year_month(x))
```

```python
cluster_data = pd.read_csv('clustering.csv')
cluster_data
```

```python
data1['cluster'] = 0
```

```python
n=0
for i in set(cluster_data['cluster']):
    n+=1
    temp = cluster_data[cluster_data['cluster']==i]
    for j in set(temp['행정동명']):
        data1['cluster'][(data1['emd_nm']==j)] = n
        
data1.head()
```

```python
temp1 = data1.copy()
temp1 = temp1.groupby(['cluster', 'year_month']).sum().reset_index()
```

```python
temp1['em_g'] = np.log(temp1['em_g'])
```

```python
temp2 = data2.copy()
temp2_resd = temp2.groupby(['emd_cluster', 'age_cluster_resd', 'age_time_cluster_resd', 'year_month']).sum().reset_index()
temp2_work = temp2.groupby(['emd_cluster', 'age_cluster_work', 'age_time_cluster_work', 'year_month']).sum().reset_index()
temp2_visit = temp2.groupby(['emd_cluster', 'age_cluster_visit', 'age_time_cluster_visit', 'year_month']).sum().reset_index()
```

```python
temp2_resd = temp2_resd[['emd_cluster', 'age_cluster_resd', 'age_time_cluster_resd', 'year_month', 'resd_pop_cnt']]
temp2_work = temp2_work[['emd_cluster', 'age_cluster_work', 'age_time_cluster_work', 'year_month', 'work_pop_cnt']]
temp2_visit = temp2_visit[['emd_cluster', 'age_cluster_visit', 'age_time_cluster_visit', 'year_month', 'visit_pop_cnt']]
```

```python
method_resd = pd.DataFrame()
method_work = pd.DataFrame()
method_visit = pd.DataFrame()
```

```python
from statsmodels.tsa.stattools import adfuller
temp_cluster = []
temp_diff = []
temp_log = []
temp_log_diff = []
for i in tqdm(set(temp2_resd['emd_cluster'])):
    temp_1 = temp2_resd[temp2_resd['emd_cluster']==i]
    for j in set(temp_1['age_cluster_resd']):
        temp_2 = temp_1[temp_1['age_cluster_resd']== j]
        for k in set(temp_2['age_time_cluster_resd']):
            temp_3 = temp_2['resd_pop_cnt'][temp_2['age_time_cluster_resd'] == k]
            temp_diff_result = adfuller(temp_3.diff()[1:])
            temp_result = adfuller(np.log(temp_3))
            temp_result_diff = adfuller(np.log(temp_3).diff()[1:])
            print("diff")
            print("p-values = ", round(temp_diff_result[1],3), " : emd_cluster", i, ' age_cluster', j, ' time_cluster', k)
            print("log")
            print("p-values = ", round(temp_result[1],3), " : emd_cluster", i, ' age_cluster', j, ' time_cluster', k)
            print("log.diff")
            print("p-values = ", round(temp_result_diff[1],3), " : emd_cluster", i, ' age_cluster', j, ' time_cluster', k)
            temp_cluster.append(str(i) + '-' + str(j) + '-' + str(k))
            if (temp_diff_result[1] < 0.05):
                temp_diff.append(temp_diff_result[1])
            else:
                temp_diff.append(np.nan)
            if (temp_result[1] < 0.05):
                temp_log.append(temp_result[1])
            else:
                temp_log.append(np.nan)
            if (temp_result_diff[1] < 0.05):
                temp_log_diff.append(temp_result_diff[1])
            else:
                temp_log_diff.append(np.nan)
```

```python
method_resd['cluster'] = temp_cluster
method_resd['diff'] = temp_diff
method_resd['log'] = temp_log
method_resd['log_diff'] = temp_log_diff
method_resd['method'] = method_resd[['diff', 'log', 'log_diff']].idxmin(axis=1)
```

```python
temp1 = data1.groupby(['cluster', 'year_month']).sum().reset_index()
temp_cluster = []
temp_diff = []
temp_log = []
temp_log_diff = []
for i in tqdm(set(temp1['cluster'])):
    temp_1 = temp1['em_g'][temp1['cluster'] == i]
    temp_diff_result = adfuller(temp_1.diff()[1:])
    temp_log_result = adfuller(np.log(temp_1))
    temp_log_diff_result = adfuller(np.log(temp_1).diff()[1:])
    temp_cluster.append(i)
    if (temp_diff_result[1] < 0.05):
        temp_diff.append(temp_diff_result[1])
    else:
        temp_diff.append(np.nan)
    if (temp_log_result[1] < 0.05):
        temp_log.append(temp_log_result[1])
    else:
        temp_log.append(np.nan)
    if (temp_log_diff_result[1] < 0.05):
        temp_log_diff.append(temp_log_diff_result[1])
    else:
        temp_log_diff.append(np.nan)
    
method_emg = pd.DataFrame()
method_emg['cluster'] = temp_cluster
method_emg['diff'] = temp_diff
method_emg['log'] = temp_log
method_emg['log_diff'] = temp_log_diff
method_emg['method'] = method_emg[['diff', 'log', 'log_diff']].idxmin(axis=1)
```

```python
temp_cluster = []
temp_result = []
n=0
from statsmodels.tsa.stattools import coint
for i in tqdm(set(temp2_resd['emd_cluster'])):
    temp_1 = temp2_resd[temp2_resd['emd_cluster'] == i]
    temp_emg = temp1[['em_g','year_month']][temp1['cluster'] == i].set_index('year_month')
    for j in set(temp_1['age_cluster_resd']):
        temp_2 = temp_1[temp_1['age_cluster_resd'] == j]
        for k in set(temp_2['age_time_cluster_resd']):
            temp_3 = temp_2[temp_2['age_time_cluster_resd'] == k].set_index('year_month')
            temp_method = method_resd['method'].iloc[n]
            temp_data = pd.merge(temp_emg, temp_3, on='year_month')
            temp_data['em_g'] = temp_data['em_g'].diff()
            
            
            if temp_method == 'diff':
                temp_data['resd_pop_cnt'] = temp_data['resd_pop_cnt'].diff().shift(1)
                temp_data = temp_data.dropna()
                _, p_value, _ = coint(temp_data['resd_pop_cnt'], temp_data['em_g'])
                if (p_value > 0.05):
                    temp_result.append(np.nan)
                else:
                    temp_result.append(p_value)
            elif temp_method == 'log':
                temp_data['resd_pop_cnt'] = np.log(temp_data['resd_pop_cnt']).shift(2)
                temp_data = temp_data.dropna()
                _, p_value, _ = coint(temp_data['resd_pop_cnt'], temp_data['em_g'])
                if (p_value > 0.05):
                    temp_result.append(np.nan)
                else:
                    temp_result.append(p_value)
            elif temp_method == 'log_diff':
                temp_data['resd_pop_cnt'] = np.log(temp_data['resd_pop_cnt']).diff().shift(1)
                temp_data = temp_data.dropna()
                _, p_value, _ = coint(temp_data['resd_pop_cnt'], temp_data['em_g'])
                if (p_value > 0.05):
                    temp_result.append(np.nan)
                else:
                    temp_result.append(p_value)
            else:
                temp_result.append(np.nan)
                
            n += 1
            temp_cluster.append(str(i)+'-'+str(j)+'-'+str(k))
coint_resd = pd.DataFrame()
coint_resd['p_value'] = temp_result
coint_resd['cluster'] = temp_cluster
```
